```python
import pandas as pd
weather = pd.read_csv('weather.csv')
```

```python
abb = {"D-backs" : "ARI", "Braves" : "ATL", "Orioles" : "BAL", "Red Sox" : "BOS", "Cubs" : "CHC", "White Sox" : "CHW", "Reds" : "CIN", "Indians" : "CLE", "Rockies" : "COL", "Tigers" : "DET", "Marlins" : "FLA", "Astros" : "HOU", "Royals" : "KAN", "Angels" : "LAA", "Dodgers" : "LAD", "Brewers" : "MIL", "Twins" : "MIN", "Mets" : "NYM", "Yankees" : "NYY", "Athletics" : "OAK", "Phillies" : "PHI", "Pirates" : "PIT", "Padres" : "SD", "Giants" : "SF", "Mariners" : "SEA", "Cardinals" : "STL", "Rays" : "TB", "Rangers" : "TEX", "Blue Jays"  : "TOR", "Nationals" : "WAS"}
for i in range(len(weather)):
    for word, initial in abb.items():
        weather.at[i, 'Home'] = weather.at[i, 'Home'].replace(word, initial)
        weather.at[i, 'Away'] = weather.at[i, 'Away'].replace(word, initial)
weather.head()
```

```python
ven = {"Globe Life Park in Arlington" : "Globe Life Field", "US Cellular Field" : "Guaranteed Rate Field", "Oco Coliseum" : "RingCentral Coliseum", "Angel Stadium of Anaheim" : "Angel Stadium", "AT&T Park" : "Oracle Park", "Turner Field" : "Truist Park", "Safeco Field" : "T-Mobile Park", "Oakland Coliseum" : "RingCentral Coliseum", "SunTrust Park" : "Truist Park"}
for i in range(len(weather)):
    for word, rep in ven.items():
        weather.at[i, 'Venue'] = weather.at[i, 'Venue'].replace(word, rep)
weather.head()
```

```python
weather = weather.rename(columns = {'Home' : 'Team 1', 'Away' : 'Team 2', 'Home Score' : 'Team 1 Score', 'Away Score' : 'Team 2 Score'})
weather.head()
```

```python
weather.to_csv('weather2.csv')
```
