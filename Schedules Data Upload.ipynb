```python
import datetime as dt
df = df1.drop(columns= [df1.columns[0]])
df["Date"] = [x[0:10] for x in df["Date"]]
df['Date'] = df['Date'].apply(lambda x: 
                                    dt.datetime.strptime(x,'%Y-%m-%d'))
```
